```python
OCdf = pd.read_csv("OCMergeLoc.csv")
```

```python
OCdf.head()
```

```python
# Add data from schools as layer for points (yes = blue, no = blue)
df = OCdf.dropna()

yes_df = df[df['onlineFlg'].astype(str) == 'True']
yes_df = yes_df[['lat', 'long']]

no_df = df[df['onlineFlg'].astype(str) == 'False']
no_df = no_df[['lat', 'long']]

yes_layer = gmaps.symbol_layer(
    yes_df, fill_color='rgba(1, 50, 67, 1)',
    stroke_color='rgba(1, 50, 67, 1)', scale=2
)

no_layer = gmaps.symbol_layer(
    no_df, fill_color='rgba(233,30, 99, 0.7)',
    stroke_color='rgba(233,30, 99, 0.7)', scale=2
)

# Base map
OC_coord = (33.7175, -117.8311) # find coordinates for your geographical areas
figure_layout = {
    'width': '800px', # you can tweak these around
    'height': '800px',
    'border': '1px solid black',
    'padding': '1px'
}

fig = gmaps.figure(center=OC_coord, zoom_level=11, layout=figure_layout)
fig.add_layer(yes_layer)
fig.add_layer(no_layer)
fig
```
